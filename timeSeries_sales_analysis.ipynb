Time Series Sales Analysis

```python

import pandas as pd
import numpy as np

# Generate the time series data
dates = pd.date_range('2024-01-01', periods=12, freq='ME')
sales = [200, 220, 250, 270, 300, 320, 350, 360, 380, 400, 420, 450]
df = pd.DataFrame({'Date': dates, 'Sales': sales})

print("Original DataFrame:")
print(df)
print("" + "="*60 + "")

# Task 1: Convert Date to datetime index
print("Task 1: Convert Date to Datetime Index")
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
print("DataFrame with Date as index:")
print(df)
print("" + "="*60 + "")

# Task 2: Compute monthly percentage growth
print("Task 2: Monthly Percentage Growth")
df['Monthly_Growth_%'] = df['Sales'].pct_change() * 100
print("DataFrame with Monthly Growth:")
print(df)
print("" + "="*60 + "")

# Task 3: Calculate 3-month rolling average sales
print("Task 3: 3-Month Rolling Average Sales")
df['Rolling_Avg_3M'] = df['Sales'].rolling(window=3).mean()
print("DataFrame with Rolling Average:")
print(df)
print("" + "="*60 + "")

# Task 4: Identify the month with highest growth rate
print("Task 4: Month with Highest Growth Rate")
max_growth_idx = df['Monthly_Growth_%'].idxmax()
max_growth_value = df['Monthly_Growth_%'].max()
max_growth_sales = df.loc[max_growth_idx, 'Sales']
print("Month with Highest Growth:", max_growth_idx.strftime('%B %Y'))
print("\nGrowth Rate:", max_growth_value)
print("\nSales in that month:", max_growth_sales)
print("" + "="*60 + "")

print("Summary Statistics:")
print("Average Monthly Growth:", df['Monthly_Growth_%'].mean())
print("Total Sales Growth (Jan to Dec):", ((df['Sales'].iloc[-1] - df['Sales'].iloc[0]) / df['Sales'].iloc[0] * 100))

```
