Sales Data Analysis

```python

import numpy as np

# Daily sales data for 10 days
sales = np.array([120, 150, 180, 200, 170, 160, 210, 220, 190, 175])

print("Original Sales Data:")
print(sales)
print()

# Task 1: Compute total sales
total_sales = np.sum(sales)
print("Task 1: Total Sales:")
print(total_sales)

# Task 2: Compute mean and standard deviation
mean_sales = np.mean(sales)
std_sales = np.std(sales)
print("\nTask 2: Mean and Standard Deviation:\n")
print("Mean:", mean_sales)
print("\nStandard Deviation:", std_sales)

# Task 3: Identify days where sales exceeded the average
days_above_avg = np.where(sales > mean_sales)[0] + 1  # +1 to make it 1-indexed
sales_above_avg = sales[sales > mean_sales]
print("\nTask 3: Days where sales exceeded the average:\n")
print("Days:", days_above_avg)
print("\nSales on those days:", sales_above_avg)

# Task 4: Normalize the sales using Min-Max scaling
# Formula: (x - min) / (max - min)
min_sales = np.min(sales)
max_sales = np.max(sales)
normalized_sales = (sales - min_sales) / (max_sales - min_sales)
print("\nTask 4: Normalized Sales (Min-Max Scaling):")
print(normalized_sales)



```
