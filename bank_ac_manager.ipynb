```python

# Bank Account Manager

# Variables to store account information
account_holder = "John Doe"
account_number = "ACC123456"
balance = 1000.0

# Function to display balance
def balance_inquiry():
    print("Account Holder:",account_holder)
    print("\nAccount Number:",account_number)
    print("\nCurrent Balance:",balance)


# Function to deposit money
def deposit(amount):
    global balance
    if amount <= 0:
        print("Error: Deposit amount must be positive!")
        return False

    balance = balance + amount
    print("Successfully deposited amount:",amount)
    print("\nNew balance:",balance)
    return True

# Function to withdraw money
def withdraw(amount):
    global balance

    if amount <= 0:
        print("Error: Withdrawal amount must be positive!")
        return False

    # Check for sufficient balance before withdrawal
    if amount > balance:
        print("Error: Insufficient balance!")
        print("\nCurrent balance:",balance)
        print("\nRequested withdrawal:",amount)
        return False

    balance = balance - amount
    print("Successfully withdrew amount:", amount)
    print("\nNew balance:", balance)
    return True

# Main program
print("Welcome to Bank Account Manager")

# Display initial balance
print("Initial Account Status:")
balance_inquiry()

# Perform deposit
print("--- Deposit Operation ---")
deposit(500.0)

# Perform withdrawal with sufficient balance
print("--- Withdrawal Operation ---")
withdraw(300.0)

# Try withdrawal with insufficient balance
print("--- Insufficient Balance Test ---")
withdraw(2000.0)

# Final balance
print("Final Account Status:")
balance_inquiry()

# Interactive menu
print("--- Interactive Mode ---")
while True:
    print("Bank Account Menu:")
    print("1. Deposit")
    print("2. Withdraw")
    print("3. Balance Inquiry")
    print("4. Exit")

    choice = input("Enter your choice (1-4): ")

    if choice == '1':
        amount = float(input("Enter deposit amount: $"))
        deposit(amount)

    elif choice == '2':
        amount = float(input("Enter withdrawal amount: $"))
        withdraw(amount)

    elif choice == '3':
        balance_inquiry()

    elif choice == '4':
        print("Thank you for using Bank Account Manager!")
        print("Have a Goodday!")
        break

    else:
        print("Error: Invalid choice! Please select 1-4.")
```
