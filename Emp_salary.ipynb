 Employee Salary Adjustment


```python

import numpy as np

# Create the salaries array
salaries = np.array([45000, 52000, 61000, 48000, 75000])

print("Original Salaries:")
print(salaries)
print()

# Task 1: Increase all salaries by 10%
salaries_increased = (salaries * 1.10).astype(int)
print("Task 1: Salaries after 10% increase:\n")
print(salaries_increased)


# Task 2: Find the average salary
average_salary = np.mean(salaries_increased)
print("\nTask 2: Average salary after increase:")
print(average_salary)


# Task 3: Find the maximum and minimum salary
max_salary = np.max(salaries_increased)
min_salary = np.min(salaries_increased)
print("\nTask 3: Maximum and Minimum salaries:\n")
print("Maximum:",max_salary)
print("\nMinimum:",min_salary)

# Task 4: Print salaries greater than 60000
salaries_above_60k = salaries_increased[salaries_increased > 60000]
print("\nTask 4: Salaries greater than $60,000:")
print(salaries_above_60k)

```
