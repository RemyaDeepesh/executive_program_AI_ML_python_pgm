Sales Performance Analysis

```python

import pandas as pd
import numpy as np

# Create the sales dataset
sales_data = {
    'Region': ['North', 'South', 'East', 'West', 'North', 'South', 'East', 'West'],
    'Salesperson': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H'],
    'Sales': [12000, 15000, 13000, 17000, 16000, 14000, 15500, 16500]
}

df = pd.DataFrame(sales_data)
print("Original Sales Dataset:")
print(df)
print("" + "="*60 + "")

# Task 1: Find total sales per region
print("Task 1: Total Sales per Region")
total_sales_per_region = df.groupby('Region')['Sales'].sum()
print(total_sales_per_region)
print("" + "="*60 + "")

# Task 2: Identify the top salesperson
print("Task 2: Top Salesperson")
top_salesperson_idx = df['Sales'].idxmax()
top_salesperson = df.loc[top_salesperson_idx]
print("Top Salesperson:", top_salesperson['Salesperson'])
print("\nRegion:", top_salesperson['Region'])
print("\nSales:", top_salesperson['Sales'])
print("" + "="*60 + "")

# Task 3: Compute average sales across regions
print("Task 3: Average Sales Across Regions")
avg_sales_per_region = df.groupby('Region')['Sales'].mean()
print(avg_sales_per_region)
print("\nOverall Average Sales:", df['Sales'].mean())
print("" + "="*60 + "")

# Task 4: Sort the dataset by sales descending
print("Task 4: Dataset Sorted by Sales (Descending)")
df_sorted = df.sort_values(by='Sales', ascending=False)
print(df_sorted)

```
