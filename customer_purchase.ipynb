Customer Purchase Analysis

```python

import pandas as pd
import numpy as np

# Create the customer purchase dataset
data = {
    'CustomerID': [101, 102, 103, 104, 105, 106, 107],
    'Age': [25, 45, 35, 50, 23, 40, 60],
    'PurchaseAmount': [250, 600, 400, 800, 150, 700, 900],
    'City': ['Delhi', 'Mumbai', 'Delhi', 'Chennai', 'Mumbai', 'Delhi', 'Chennai']
}

df = pd.DataFrame(data)
print("Original Customer Purchase Dataset:")
print(df)
print("" + "="*60 + "")

# Task 1: Find average purchase amount by city
print("Task 1: Average Purchase Amount by City")
avg_purchase_by_city = df.groupby('City')['PurchaseAmount'].mean()
print(avg_purchase_by_city)
print("" + "="*60 + "")

# Task 2: Identify customers with purchase above overall average
print("Task 2: Customers with Purchase Above Overall Average")
overall_avg = df['PurchaseAmount'].mean()
print("\nOverall Average Purchase:", overall_avg)
customers_above_avg = df[df['PurchaseAmount'] > overall_avg]
print("\nCustomers above average:")
print(customers_above_avg)
print("" + "="*60 + "")

# Task 3: Find the city generating highest revenue
print("Task 3: City Generating Highest Revenue")
total_revenue_by_city = df.groupby('City')['PurchaseAmount'].sum()
print("\nTotal Revenue by City:")
print(total_revenue_by_city)
highest_revenue_city = total_revenue_by_city.idxmax()
highest_revenue = total_revenue_by_city.max()
print("Highest Revenue City:", highest_revenue_city," with", highest_revenue)
print("" + "="*60 + "")

# Task 4: Create Age_Group column
print("Task 4: Create Age_Group Column")
def categorize_age(age):
    if age < 30:
        return 'Young'
    elif age <= 50:
        return 'Adult'
    else:
        return 'Senior'

df['Age_Group'] = df['Age'].apply(categorize_age)
print("DataFrame with Age_Group:")
print(df)
print("" + "="*60 + "")

# Task 5: Find average purchase per age group
print("Task 5: Average Purchase per Age Group")
avg_purchase_by_age_group = df.groupby('Age_Group')['PurchaseAmount'].mean()
print(avg_purchase_by_age_group)
print("Detailed breakdown:")
for age_group in ['Young', 'Adult', 'Senior']:
    if age_group in avg_purchase_by_age_group.index:
        count = len(df[df['Age_Group'] == age_group])
        avg = avg_purchase_by_age_group[age_group]
        print(age_group ,":", avg, "Count:" , count, "customers")

```
