Employee Dataset Exploration using Dataframe

```python

import pandas as pd
import numpy as np

# Task 1: Create a DataFrame from the dictionary
data = {
    'Employee': ['A', 'B', 'C', 'D', 'E', 'F'],
    'Department': ['HR', 'IT', 'IT', 'Finance', 'HR', 'Finance'],
    'Salary': [50000, 65000, 70000, 72000, 52000, 68000],
    'Experience': [2, 5, 7, 6, 3, 4]
}

df = pd.DataFrame(data)
print("Task 1: DataFrame Created")
print(df)
print("" + "="*60 + "")

# Task 2: Display basic statistics
print("Task 2: Basic Statistics of the Dataset")
print(df.describe())
print("" + "="*60 + "")

# Task 3: Find average salary per department
print("Task 3: Average Salary per Department")
avg_salary_per_dept = df.groupby('Department')['Salary'].mean()
print(avg_salary_per_dept)
print("" + "="*60 + "")

# Task 4: Identify employees with experience greater than 4 years
print("Task 4: Employees with Experience > 4 years")
experienced_employees = df[df['Experience'] > 4]
print(experienced_employees)
print("" + "="*60 + "")

# Task 5: Add a new column Bonus = 10% of salary
df['Bonus'] = df['Salary'] * 0.10
print("Task 5: DataFrame with Bonus Column (10% of Salary)")
print(df)

```
