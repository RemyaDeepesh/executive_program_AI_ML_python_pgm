```python

import pandas as pd
import numpy as np
from scipy import stats

# ─────────────────────────────────────────────────────────────
# Load Dataset
# ─────────────────────────────────────────────────────────────
df = pd.read_csv("/content/drive/MyDrive/ExecutiveProgramInAdvanced-AI ML-IIT Palakkad-Feb2026/StudentsPerformance.csv")
df.columns = df.columns.str.strip()

math    = df["math score"]
reading = df["reading score"]
writing = df["writing score"]

print("=" * 65)
print("       STUDENT PERFORMANCE DATA ANALYSIS")
print("=" * 65)

# ─────────────────────────────────────────────────────────────
# TASK 1: Mean, Median, Mode
# ─────────────────────────────────────────────────────────────
print(" TASK 1: Mean, Median & Mode")
print("-" * 65)
for name, col in [("Math", math), ("Reading", reading), ("Writing", writing)]:
    print(f"{name} Score:")
    print(f"    Mean   : {col.mean():.2f}")
    print(f"    Median : {col.median():.2f}")
    print(f"    Mode   : {col.mode().values}")

# ─────────────────────────────────────────────────────────────
# TASK 2: Range, Variance, Standard Deviation
# ─────────────────────────────────────────────────────────────
print("TASK 2: Range, Variance & Standard Deviation")
print("-" * 65)
for name, col in [("Math", math), ("Reading", reading), ("Writing", writing)]:
    print(f"{name} Score:")
    print(f"    Range    : {col.max() - col.min()}")
    print(f"    Variance : {col.var():.4f}")
    print(f"    Std Dev  : {col.std():.4f}")

# ─────────────────────────────────────────────────────────────
# TASK 3: Math Score Stats by Parental Education Level
# ─────────────────────────────────────────────────────────────
print("TASK 3: Math Score Stats by Parental Education Level")
print("-" * 65)
edu_order = [
    "some high school", "high school", "some college",
    "associate's degree", "bachelor's degree", "master's degree"
]
for edu in edu_order:
    subset = df[df["parental level of education"] == edu]["math score"]
    print(f"{edu.title()} (n={len(subset)}):")
    print(f"    Mean   : {subset.mean():.2f}")
    print(f"    Median : {subset.median():.2f}")
    print(f"    Mode   : {subset.mode().values}")

# ─────────────────────────────────────────────────────────────
# TASK 4: Mean Scores – Test Prep vs No Test Prep
# ─────────────────────────────────────────────────────────────
print("TASK 4: Mean Scores – Test Prep vs No Test Prep")
print("-" * 65)
for group, label in [("completed", "Completed"), ("none", "None")]:
    sub = df[df["test preparation course"] == group]
    print(f"Test Prep: {label} (n={len(sub)})")
    print(f"    Math    Mean : {sub['math score'].mean():.2f}")
    print(f"    Reading Mean : {sub['reading score'].mean():.2f}")
    print(f"    Writing Mean : {sub['writing score'].mean():.2f}")

# ─────────────────────────────────────────────────────────────
# TASK 5: Interquartile Range (IQR)
# ─────────────────────────────────────────────────────────────
print("TASK 5: Interquartile Range (IQR)")
print("-" * 65)
for name, col in [("Math", math), ("Reading", reading), ("Writing", writing)]:
    q1  = col.quantile(0.25)
    q3  = col.quantile(0.75)
    iqr = q3 - q1
    print(f"{name} Score:")
    print(f"    Q1  : {q1}")
    print(f"    Q3  : {q3}")
    print(f"    IQR : {iqr}")

# ─────────────────────────────────────────────────────────────
# TASK 6: Math Score Std Dev by Lunch Type
# ─────────────────────────────────────────────────────────────
print("TASK 6: Math Score Std Dev by Lunch Type")
print("-" * 65)
for lunch in ["free/reduced", "standard"]:
    sub = df[df["lunch"] == lunch]["math score"]
    print(f" Lunch: {lunch.title()} (n={len(sub)})")
    print(f"    Mean    : {sub.mean():.2f}")
    print(f"    Std Dev : {sub.std():.4f}")

# ─────────────────────────────────────────────────────────────
# TASK 7: Correlation Coefficients
# ─────────────────────────────────────────────────────────────
print("TASK 7: Correlation Coefficients")
print("-" * 65)
r_math_reading    = math.corr(reading)
r_reading_writing = reading.corr(writing)
print(f" Math vs Reading    : r = {r_math_reading:.4f}")
print(f"  Reading vs Writing : r = {r_reading_writing:.4f}")

# ─────────────────────────────────────────────────────────────
# TASK 8: Interpretation of Correlations
# ─────────────────────────────────────────────────────────────
print("TASK 8: Interpretation of Correlations")
print("-" * 65)

def interpret_correlation(r):
    abs_r = abs(r)
    direction = "positive" if r > 0 else "negative"
    if abs_r >= 0.9:
        strength = "Very Strong"
    elif abs_r >= 0.7:
        strength = "Strong"
    elif abs_r >= 0.5:
        strength = "Moderate"
    elif abs_r >= 0.3:
        strength = "Weak"
    else:
        strength = "Very Weak / Negligible"
    return f"{strength} {direction} correlation"

print(f"Math vs Reading    (r={r_math_reading:.4f}) → {interpret_correlation(r_math_reading)}")
print(f"Students who score well in math tend to also score well in reading.")
print(f"Likely reflects overall academic ability and study habits.")

print(f"Reading vs Writing (r={r_reading_writing:.4f}) → {interpret_correlation(r_reading_writing)}")
print(f"Reading and writing are deeply linked literacy skills.")
print(f"Proficiency in one almost always accompanies proficiency in the other.")

# ─────────────────────────────────────────────────────────────
# TASK 9: Covariance vs Correlation
# ─────────────────────────────────────────────────────────────
print("TASK 9: Covariance – Math vs Reading")
print("-" * 65)
cov_math_reading = np.cov(math, reading, ddof=1)[0][1]
print(f"Covariance  (Math, Reading) : {cov_math_reading:.4f}")
print(f"  Correlation (Math, Reading) : {r_math_reading:.4f}")
print(f"Covariance is in score² units — hard to interpret in isolation.")
print(f"Correlation normalises it to [-1, 1], making strength & direction clear.")
print(f"Both confirm a strong positive relationship between math and reading.")

# ─────────────────────────────────────────────────────────────
# TASK 10: One-Sample Z-Test (H₀: μ = 50)
# ─────────────────────────────────────────────────────────────
print("TASK 10: One-Sample Z-Test (National Average = 50)")
print("-" * 65)
mu0    = 50
n      = len(math)
x_bar  = math.mean()
sigma  = math.std()
z_stat = (x_bar - mu0) / (sigma / np.sqrt(n))
p_val_z = 2 * (1 - stats.norm.cdf(abs(z_stat)))

print(f"H₀: μ = {mu0}  |  H₁: μ ≠ {mu0}")
print(f"  Sample Mean  : {x_bar:.4f}")
print(f"  Std Dev      : {sigma:.4f}")
print(f"  n            : {n}")
print(f"  Z-Statistic  : {z_stat:.4f}")
print(f"  P-Value      : {p_val_z:.6f}")
print(f"  Decision     : {'Reject H₀ — Significant difference from national average' if p_val_z < 0.05 else 'Fail to Reject H₀'}")

# ─────────────────────────────────────────────────────────────
# TASK 11: Two-Sample T-Test (Reading by Lunch Type)
# ─────────────────────────────────────────────────────────────
print("TASK 11: Two-Sample T-Test – Reading Score by Lunch Type")
print("-" * 65)
free_reading = df[df["lunch"] == "free/reduced"]["reading score"]
std_reading  = df[df["lunch"] == "standard"]["reading score"]
t_stat, p_val_t = stats.ttest_ind(free_reading, std_reading, equal_var=False)

print(f"H₀: No difference in reading scores between lunch groups")
print(f"  Free/Reduced — Mean: {free_reading.mean():.2f}, n={len(free_reading)}")
print(f"  Standard     — Mean: {std_reading.mean():.2f}, n={len(std_reading)}")
print(f"  T-Statistic  : {t_stat:.4f}")
print(f"  P-Value      : {p_val_t:.6f}")
print(f"  Decision     : {'Reject H₀ — Significant difference in reading scores' if p_val_t < 0.05 else 'Fail to Reject H₀'}")

# ─────────────────────────────────────────────────────────────
# TASK 12: F-Test (Math Variance by Test Prep)
# ─────────────────────────────────────────────────────────────
print("TASK 12: F-Test – Math Score Variance by Test Prep")
print("-" * 65)
prep_math    = df[df["test preparation course"] == "completed"]["math score"]
no_prep_math = df[df["test preparation course"] == "none"]["math score"]
var1   = prep_math.var(ddof=1)
var2   = no_prep_math.var(ddof=1)
f_stat = var1 / var2
df1    = len(prep_math) - 1
df2    = len(no_prep_math) - 1
p_val_f = 2 * min(stats.f.cdf(f_stat, df1, df2), 1 - stats.f.cdf(f_stat, df1, df2))

print(f"H₀: Variances are equal between test prep groups")
print(f"  Completed — Variance: {var1:.4f}, n={len(prep_math)}")
print(f"  None      — Variance: {var2:.4f}, n={len(no_prep_math)}")
print(f"  F-Statistic : {f_stat:.4f}")
print(f"  df1={df1}, df2={df2}")
print(f"  P-Value     : {p_val_f:.6f}")
print(f"  Decision    : {'Reject H₀ — Variances differ significantly' if p_val_f < 0.05 else 'Fail to Reject H₀ — No significant difference in variances'}")

print("" + "=" * 65)
print("                   ANALYSIS COMPLETE")
print("=" * 65)

```
