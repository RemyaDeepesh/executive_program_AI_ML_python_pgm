```python

# Weather Data Analyzer

# Variables to store weather data for a week
temperatures = [28, 32, 30, 25, 27, 31, 29]  # in Celsius
humidity = [65, 70, 68, 80, 75, 60, 72]      # in percentage
days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

# Function to calculate average temperature
def calculate_average_temperature():
    total = 0
    for temp in temperatures:
        total = total + temp
    average = total / len(temperatures)
    return average

# Function to calculate average humidity
def calculate_average_humidity():
    total = 0
    for hum in humidity:
        total = total + hum
    average = total / len(humidity)
    return average

# Function to determine weather condition based on temperature and humidity
def determine_weather_condition(temp, hum):
    if temp > 30 and hum < 70:
        return "Sunny"
    elif temp > 25 and hum >= 70:
        return "Humid"
    elif temp <= 25 and hum >= 75:
        return "Rainy"
    elif temp <= 20:
        return "Cold"
    else:
        return "Pleasant"

# Function to display daily weather report
def display_daily_weather():
    print("DAILY WEATHER REPORT")

    for i in range(len(days)):
        condition = determine_weather_condition(temperatures[i], humidity[i])
        print(f"{days[i]:12} | Temp: {temperatures[i]}°C | Humidity: {humidity[i]}% | {condition}")


# Function to find hottest and coldest days
def find_extreme_temperatures():
    max_temp = temperatures[0]
    min_temp = temperatures[0]
    max_day = days[0]
    min_day = days[0]

    for i in range(len(temperatures)):
        if temperatures[i] > max_temp:
            max_temp = temperatures[i]
            max_day = days[i]
        if temperatures[i] < min_temp:
            min_temp = temperatures[i]
            min_day = days[i]

    print(f"Hottest Day: {max_day} with {max_temp}°C")
    print(f"Coldest Day: {min_day} with {min_temp}°C")

# Function to print weather summary
def print_weather_summary():
    avg_temp = calculate_average_temperature()
    avg_hum = calculate_average_humidity()

    print("WEATHER ANALYSIS SUMMARY")
    print("\nRegion: Sample City")
    print("\nAnalysis Period: 7 Days")
    print("\nAverage Temperature:",avg_temp)
    print("\nAverage Humidity:",avg_hum)
    print()
    find_extreme_temperatures()

# Main Program
print("Welcome to Weather Data Analyzer")

# Display daily weather report
display_daily_weather()

# Display weather summary
print_weather_summary()

# Interactive menu
print("--- Interactive Mode ---")
while True:
    print("Weather Analyzer Menu:")
    print("1. View Daily Weather Report")
    print("2. View Weather Summary")
    print("3. Check Specific Day Weather")
    print("4. Add New Day Data")
    print("5. Exit")

    choice = input("Enter your choice (1-5): ")

    if choice == '1':
        display_daily_weather()

    elif choice == '2':
        print_weather_summary()

    elif choice == '3':
        print("Available days:")
        for i in range(len(days)):
            print(f"{i+1}. {days[i]}")
        day_num = int(input("Enter day number (1-7): "))

        if 1 <= day_num <= len(days):
            index = day_num - 1
            condition = determine_weather_condition(temperatures[index], humidity[index])
            print(f"{days[index]} Weather:")
            print(f"Temperature: {temperatures[index]}°C")
            print(f"Humidity: {humidity[index]}%")
            print(f"Condition: {condition}")
        else:
            print("Error: Invalid day number!")

    elif choice == '4':
        day_name = input("Enter day name: ")
        temp = float(input("Enter temperature (°C): "))
        hum = float(input("Enter humidity (%): "))

        days.append(day_name)
        temperatures.append(temp)
        humidity.append(hum)

        print(f"Successfully added weather data for {day_name}!")

    elif choice == '5':
        print("Thank you for using Weather Data Analyzer")
        print("Goodbye")
        break

    else:
        print("Error: Invalid choice! Please select 1-5.")
```
