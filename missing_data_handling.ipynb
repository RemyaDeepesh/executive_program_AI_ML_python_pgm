Missing Data Handling

```python

import pandas as pd
import numpy as np

# Create the dataset with missing values
data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda', 'David'],
    'Age': [28, np.nan, 35, 32, np.nan],
    'Salary': [50000, 54000, np.nan, 58000, 60000]
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)
print("" + "="*60 + "")

# Task 1: Detect missing values
print("Task 1: Detect Missing Values")
print("\nMissing values count per column:")
print(df.isnull().sum())
print("\nMissing values (True/False):")
print(df.isnull())
print("" + "="*60 + "")

# Task 2: Replace missing Age with mean age
print("Task 2: Replace Missing Age with Mean Age")
mean_age = df['Age'].mean()
print(f"Mean Age: {mean_age:.2f}")
df['Age'] = df['Age'].fillna(mean_age)
print("\nDataFrame after filling Age:")
print(df)
print("" + "="*60 + "")

# Task 3: Replace missing Salary with median salary
print("Task 3: Replace Missing Salary with Median Salary")
median_salary = df['Salary'].median()
print("Median Salary:", median_salary)
df['Salary'] = df['Salary'].fillna(median_salary)
print("DataFrame after filling Salary:")
print(df)
print("" + "="*60 + "")

# Task 4: Remove rows with any remaining missing values
print("Task 4: Remove Rows with Any Remaining Missing Values")
print(f"Missing values after imputation:")
print(df.isnull().sum())
df_cleaned = df.dropna()
print("Rows before data cleaning:", len(df), "\nRows after data cleaning:", len(df_cleaned))
print("Final Cleaned DataFrame:")
print(df_cleaned)

```
